## ðŸŸ¨ Convert to Dictionary (1.21.2)

```python
# Create the JobPostings class in this code cell
class JobPosting:
  def __init__(self, title, company, location, salary):
    self.title = title
    self.company = company
    self.location = location
    self.salary = salary

  def to_dict(self):
    return {'title': self.title, 'company': self.company, 'location': self.location, 'salary': self.salary}
```

```python
# Code Needed
job = JobPosting('Data Scientist', 'Tech Innovations', 'New York', 120000)
jobdict = job.to_dict()
print(jobdict)
```

## ðŸŸ¥ Compare Job Salaries (1.21.3)

```python
# Create the JobPostings class in this code cell
class JobPosting:
  def __init__(self, title, company, location, salary):
    self.title = title
    self.company = company
    self.location = location
    self.salary = salary

  def compare_salary(self, job2):
    if self.salary > job2.salary:
      return self.title
    else:
      return job2.title
```

```python
# Code Needed
job1 = JobPosting('Data Scientist', 'Tech Innovations', 'New York', 120000)
job2 = JobPosting('Data Analyst', 'Data Driven Co', 'San Francisco', 100000)
print(job1.compare_salary(job2))
```
