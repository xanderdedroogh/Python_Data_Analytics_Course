```python
!pip install -U datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## üü© Job Posting Location Bar Plot (1.31.1)

```python
job_counts = df['job_via'].value_counts().head(5)
job_counts.plot(kind = 'bar')
```

---

## üü© Top Job Locations (1.31.2)

```python
top_locations = df['job_location'].value_counts().head(5)
top_locations.plot(kind='barh')
plt.xlabel('Count of Jobs')
plt.ylabel('Job Location')
plt.title('Number of Job Postings by Job Location')
```

---

## üü® Job Postings Trend Line Plot (1.31.3)

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['month'] = df['job_posted_date'].dt.to_period('M')
monthly_job_counts = df['month'].value_counts().sort_index()
monthly_job_counts.plot(kind= 'line')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.title('Trend of Job Postings Over Time')
plt.show()
```

```python
median_salary = df.groupby('job_title_short')['salary_year_avg'].median()
median_salary = median_salary.sort_values()
median_salary.plot(kind='barh')
plt.xlabel('Median Salary')
plt.ylabel('Job Title')
plt.title('Median Salary by Job Title')
plt.show()
```
