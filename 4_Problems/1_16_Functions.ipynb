## ðŸŸ¨ Salary Statistics (1.16.3)

```python
# Code Needed
def average(salaries):
  return sum(salaries)/len(salaries)

def salary_stats(salaries):
  return {'Minimum': min(salaries), 'Maximum': max(salaries), 'Average': average(salaries)}

salaries = [95000, 120000, 105000, 90000, 130000]
salary_stats(salaries)
```

## ðŸŸ¥ Job Posting Summary (1.16.4)

```python
# Code Needed
job_postings = [
    {'title': 'Data Scientist', 'location': 'New York', 'salary': 95000},
    {'title': 'Data Analyst', 'location': 'San Francisco', 'salary': 85000},
    {'title': 'Machine Learning Engineer', 'location': 'New York', 'salary': 115000}
]

def job_posting_summary(job_postings):

  number = len(job_postings)

  total_salary = 0
  for posting in job_postings:
    total_salary += posting["salary"]
  average_income = total_salary / len(job_postings)

  locations = set()
  for posting in job_postings:
    locations.add(posting["location"])
  unique_locations = list(locations)

  return {'Number of postings': number, 'Average income': average_income, 'All locations': unique_locations}
job_posting_summary(job_postings)
```
