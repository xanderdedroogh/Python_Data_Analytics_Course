## ðŸŸ¥ Determine Job Title (1.7.3)

```python
# Code Needed
job_experience = 4
sql_experience = 2
if job_experience < 2:
  if sql_experience == 0:
    position = 'Associate Analyst'
  else:
    position = 'Junior Analyst'
elif 2 <= job_experience <= 5:
  if sql_experience == 0:
    position = 'Associate Analyst'
  elif sql_experience == 1:
    position = 'Intermediate Analyst'
  else:
    position = 'Senior Analyst'
else:
  if 0 <= sql_experience <= 2:
    position = 'Data Analyst'
  else:
    position = 'Lead Analyst'
print(position)
```
